## Iterando sobre arrays

Se irera sobre np.arrays  tal cual se hace en Python, pero añadiendo el método ``nditer()``, que proporciona un iterador.

```python
import numpy as np
array = np.arange(1,10).reshape(3,3)

print(f"Two-dimensional array:\n{array}\n")
for element in np.nditer(array):
    print(element, end=' ')
```

```python
#Como es de esperar, el orden de iteración tiene que ver con la forma en la que está almacenado en memoria
print(f"Traspuesta:\n{array.T}\n")
#print(f"\nArray:\n{array}")
for element in np.nditer(array.T):
    print(element, end=' ')
```

```python
print(f"Traspuesta:\n{array.T}\n")

for element in np.nditer(array.T.copy()):
    print(element, end=' ')
```

```python
print(f"Traspuesta:\n{array.T}\n")
print(array)
for element in np.nditer(array.T,order='C'):
    print(element, end=' ')
```

### Modificaciones al iterar

Por defecto, las iteraciones sólo permiten operaciones de lectura, con lo que no es posible modificar al iterar sobre un array. No obstante, se puede cambiar esta propiedad. Además, es necesario conocer que hay que informar del final del proceso de iteración, ya que se trabaja con valores en memoria temporal. Se consigue incrustando la iteración en una sentencia with o invocando a cerrar el iterador (close).

```python
array_to_update = np.arange(1,10).reshape(3,3)

print(f'Array:\n{array_to_update}\n')

with np.nditer(array_to_update, op_flags=['readwrite']) as iterator:
    for element in iterator:
        element[...] = element**2

print(f'Updated array (squared):\n{array_to_update}')
```

```python
array_to_update = np.arange(1,10).reshape(3,3)

print(f'Array:\n{array_to_update}\n')
x=np.nditer(array_to_update, op_flags=['readwrite'])
for element in x:
        element[...] = element**3
x.close()

print(f'Updated array (squared):\n{array_to_update}')
```

**Ejercicio**: revisa la documentación para conocer con otras [alternativas para algoritmos de ordenación](https://wesmckinney.com/book/advanced-numpy#numpy_advanced_sorting). Pon ejemplos del uso de ``np.partition``, ``np.argpartition`` y ``searchsorted``.

```python
import numpy as np
#Array
array = np.array([13, 2, 7, 5, 18, 8, 14])
print(array)
print()

print("np partition")
#Cantidad de valores buscados
k = 7
# np.partition coloca los k menores a la izquierda del array
arreglo_parcial = np.partition(array, k-1)
print("Con np-partition:", arreglo_parcial)
mas_pequeños = np.sort(arreglo_parcial[:k])
print(f"k menores (k={k}):", mas_pequeños)

print()
# np.argpartition obtiene lo k mayores
print("np.argpartion")
indices_mayores = np.argpartition(-array, k-1)[:k]
print("top-k (mayores):", indices_mayores)
valores_ordenados = indices_mayores[np.argsort(array[indices_mayores])[::-1]]
print("Valores mayores ordenados:", array[valores_ordenados])

print()
# searchsorted requiere array ordenado
print("Searchsorted")
# Buscar posiciones
valores = np.array([1, 3, 4, 10])
posiciones = np.searchsorted(arreglo_parcial, valores)
print("Posiciones a modificar:", posiciones)
print("side='left':", np.searchsorted(arreglo_parcial, 7, side='left'))   # antes
print("side='right':", np.searchsorted(arreglo_parcial, 7, side='right')) # despues
```

**Ejercicio**:
- Crea un array ``datos1`` de configuración (7,4), que siga una distribución normal de media 5 y desviación 1.
- Crea ``datos2``, aleatoria de configuración (32,).
- Modifica datos2 para poder concatenarla por filas con ``datos1`` (calcula el número de columnas a partir de ``datos1.shape``).
- Parte esa concatenación en tres submatrices con el mismo número de filas, llámalas ``m1``, ``m2`` y ``m3``. Determina si ``m2`` está ordenada por filas (hazlo comparando las matrices que resultan de eliminar la primera y la última fila).
- Determina si ``m1`` está ordenada por columnas.

```python
#Crea un array datos1 de configuración (7,4), que siga una distribución normal de media 5 y desviación 1.
datos1 = 5 + 1 * np.random.randn(7, 4)
print("Configuracion", datos1.shape)

#Crea datos2, aleatoria de configuración (32,).
datos2 = np.random.random(32)
print("Configuracion", datos2.shape)

print()
#Modifica datos2 para poder concatenarla por filas con datos1 (calcula el número de columnas a partir de datos1.shape).
n_cols = datos1.shape[1]
Cont1_datos2 = datos2.reshape(-1, n_cols)
print("Calculo de datos1.shape en datos2", Cont1_datos2.shape)

#Parte esa concatenación en tres submatrices con el mismo número de filas,
#llámalas m1, m2 y m3. Determina si m2 está ordenada por filas (hazlo comparando las matrices que resultan de eliminar la primera y la última fila).
conc = np.concatenate([datos1, Cont1_datos2], axis=0)
print("tamaño de concatenacion:", conc.shape)
filas = conc.shape[0] // 3
m1 = conc[0:filas, :]
m2 = conc[filas:2*filas, :]
m3 = conc[2*filas:3*filas, :]
print("m1:", m1.shape, "| m2:", m2.shape, "| m3:", m3.shape)
#Validar si M2 esta organizado
if np.all(m2[1:, :] >= m2[:-1, :]):
    print("M2 esta ordenado")
else:
    print("M2 no esta ordenado")

#Determina si m1 está ordenada por columnas.
if np.all(m1[:, 1:] >= m1[:, :-1]):
    print("M1 si esta ordenada")
else:
    print("M1 no esta ordenada")
```

**Ejercicio**: Parte de la imagen del café. Calcula media y desviación típica de la matriz y de cada uno de los canales. En cada canal, a los que sean inferiores que la media, añádeles la desviación típica. A los que sean superiores, réstasela. Calcula y muestra la nueva imagen.


```python
import matplotlib.pylab as plt
import skimage
#Cargar imagen
imagen=skimage.data.coffee()

#Calcula media y desviación típica de la matriz y de cada uno de los canales
media = np.mean(imagen, axis=(0, 1))
desv = np.std(imagen, axis=(0, 1))
print("Media por canal (R,G,B):", media)
print("Desviación típica (R,G,B):", desv)

nueva = imagen.copy()

#En cada canal, a los que sean inferiores que la media, añádeles la desviación típica
for canal in range(3):
  for i in range(imagen.shape[0]):
    for j in range(imagen.shape[1]):
      valor = imagen[i, j, canal]
      if valor < media[canal]:
        valor = valor + desv[canal]
      elif valor > media[canal]:
        valor = valor - desv[canal]
      nueva[i, j, canal] = valor

print(nueva)

plt.imshow(nueva.astype(np.uint8))

plt.show()
```

**Ejercicio**: Haz una prueba de argsort y lexsort en un array de strings.

```python
#Arreglo
animales = np.array(["Gato", "Perro", "Salmon", "Aguila", "Leon"])
clase = np.array(["Mamifero", "Mamifero", "Pez", "Ave", "Mamifero"])
#argsort (Orden alfabetico)
indices = np.argsort(animales)
print("Índices ordenados:", indices)
print("Animales ordenados:", animales[indices])

print()
#lexsort (Clase secundaria y primaria)
orden = np.lexsort((animales, clase))
print("Orden por clase y luego animal:")
print("Clases ordenadas:", clase[orden])
print("Animales ordenados:", animales[orden])
```

**Ejercicio**: Crea un array aleatorio (3,4) y permuta columnas de forma que la primera fila quede ordenada. Repite el problema permutando filas para que quede ordenada la primera columna.

```python
# Crear un array aleatorio (3,4)
Array = np.random.randn(3, 4)
print("Array original", Array)

#argsort
orden_col = np.argsort(Array[0, :])
Array_col_ordenado = Array[:, orden_col]
print()
print("Array con primera fila ordenada", Array_col_ordenado)

#Permutar fila y columna
orden_fila = np.argsort(Array[:, 0])
Array_fila_ordenado = Array[orden_fila, :]
print()
print("Array con primera columna ordenada", Array_fila_ordenado)
```
